# Calculadora RISKO

```python
def age_opt(age):
  if age > 10 and age <= 20:
    return '10 - 20 anos', 1
  elif age > 20 and age <= 30:
    return '21 - 30 anos', 2
  elif age > 30 and age <= 40:
    return '31 - 40 anos', 3
  elif age > 40 and age <= 50:
    return '41 - 50 anos', 4
  elif age > 50 and age <= 60:
    return '51 - 60 anos', 6
  else:
    return 'Acima de 60 anos', 8

def fam_hist_opt(fam_hist):
  if fam_hist == '0':
    return 'Nenhuma história reconhecida', 1
  elif fam_hist == 1:
    return '1 parente com + de 60 anos com doença cardiovascular', 2
  elif fam_hist == 2:
    return '2 parentes com + de 60 anos com doença cardiovascular', 3
  elif fam_hist == 3:
    return '1 parente com - de 60 anos com doença cardiovascular', 4
  elif fam_hist == 4:
    return '2 parentes com - de 60 anos com doença cardiovascular', 6
  else:
    return '3 parente com - de 60 anos com doença cardiovascular', 8

def weight_opt(high, weight):
  weight_min = high * altura * 18.5
  weight_max = high * altura * 25
  if weight < weight_min - 2.3:
    return '+ de 2,3kg abaixo do peso padrão*', 0
  elif weight > weight_min - 2.3 and weight <= weight_max + 2.3:
    return 'De -2,3kg a +2,3kg do peso padrão*', 1
  elif weight > weight_max + 2.3 and weight <= weight_max + 9:
    return 'excesso de peso padrão* de 2,4kg à 9kg', 2
  elif weight > weight_max + 9 and weight <= 15.8:
    return 'excesso de peso padrão* de 9,1kg à 15,8kg', 3
  elif weight > 15.8 and weight <= 22.6:
    return 'excesso de peso padrão* de 15,9 kg à 22,6 kg', 5
  else:
    return 'excesso de peso padrão* acima de 22,6kg', 7

def smoke_opt(smoke):
  if smoke == 0:
    return 'Não Fumante', 0
  elif smoke > 0 and smoke <= 10:
    return '10 cigarros ou menos por dia', 1
  elif smoke > 10 and smoke <= 20:
    return 'De 11 à 20 cigarros por dia', 2
  elif smoke > 20 and smoke <= 30:
    return 'De 21 à 30 cigarros por dia', 4
  elif smoke > 30 and smoke <= 40:
    return 'De 31 à 40 cigarros por dia', 6
  else:
    return 'Acima de 40 cigarros por dia', 10

def fitness_opt(fitness):
  if fitness == 'intenso + intenso':
    return 'Esforço Ocupacional e recreacional intenso', 1
  elif fitness == 'moderado + moderado':
    return 'Esforço Ocupacional e recreacional moderado', 2
  elif fitness == 'sedentário + intenso':
    return 'Trabalho sedentário e esforço recreacional intenso', 3
  elif fitness == 'sedentário + moderado':
    return 'Trabalho sedentário e esforço recreacional moderado', 5
  elif fitness == 'sedentário + leve':
    return 'Trabalho sedentário e esforço recreacional leve', 6
  else:
    return 'Ausência completa de qualquer exercício', 8

def fat_diet_opt(fat):
  if fat == '0':
    return 'Ausência de gorduras animais ou sólidas na dieta', 1
  elif fat <= 10:
    return '10% de gordura animal ou sólida na dieta', 2
  elif fat > 10 and fat <= 20:
    return '20% de gordura animal ou sólida na dieta', 3
  elif fat > 20 and fat <= 30:
    return '30% de gordura animal ou sólida na dieta', 4
  elif fat > 30 and fat <= 40:
    return '40% de gordura animal ou sólida na dieta', 5
  elif fat > 40:
    return '50% de gordura animal ou sólida na dieta', 7

def pas_opt(pas):
  if pas < 120:
    return 'Abaixo 120 mmHg', 1
  elif pas >= 120 and pas < 140:
    return 'De 120 a 139 mmHg', 2
  elif pas >= 140 and pas < 160:
    return 'De 140 a 159 mmHg', 3
  elif pas >= 160 and pas < 180:
    return 'De 160 a 179 mmHg', 4
  elif pas >= 180 and pas < 200:
    return 'De 180 a 199 mmHg', 6
  else:
    return 'Acima de 200 mmHg', 8

def pad_opt(pad):
  if pad < 70:
    return 'Abaixo 70 mmHg', 1
  elif pad >= 70 and pad < 77:
    return 'De 71 a 76 mmHg', 2
  elif pad >= 77 and pad < 83:
    return 'De 77 a 82 mmHg', 3
  elif pad >= 83 and pad < 94:
    return 'De 83 a 93 mmHg', 5
  elif pad >= 94 and pad < 106:
    return 'De 94 a 105 mmHg', 6
  else:
    return 'Acima de 106 mmHg', 8


def registro(idade, hereditariedade, peso, altura, fumo, exercicio, gordura_dieta, pas, pad):
    idade = age_opt(idade)
    hereditariedade = fam_hist_opt(hereditariedade)
    peso = weight_opt(altura, peso)
    fumo = smoke_opt(fumo)
    exercicio = fitness_opt(exercicio)
    gordura_dieta = fat_diet_opt(gordura_dieta)
    pas = pas_opt(pas)
    pad = pad_opt(pad)
    total = idade[1] + hereditariedade[1] + peso[1] + fumo[1] + exercicio[1] + gordura_dieta[1] + pas[1] + pad[1]
    print(idade[1])
    print(hereditariedade[1])
    print(peso[1])
    print(fumo[1])
    print(exercicio[1])
    print(gordura_dieta[1])
    print(pas[1])
    print(pad[1])
    print('Total: ' + str(total))


```

## Hereditariedade:<br>

  **2+60:** 2 parentes com + de 60 anos com doença cardiovascular<br>

  **3-60:** 3 parente com - de 60 anos com doença cardiovascular<br>
## Peso (valor numérico)<br>
## Altura (valor numérico em m)<br>
## Fumo (quantidade de cigarros por dia)<br>
## Exercício:
  **intenso + intenso:** Esforço Ocupacional e recreacional intenso'<br>
  **moderado + moderado:**  Esforço Ocupacional e recreacional moderado'<br>
  **sedentário + intenso:** Trabalho sedentário e esforço recreacional intenso<br>
  **sedentário + moderado:** Trabalho sedentário e esforço recreacional moderado<br>
  **sedentário + leve:** Trabalho sedentário e esforço recreacional leve<br>
  **0:** Ausência completa de qualquer exercício<br>
## Gordura na dieta (porcentagem de gordura na dieta)<br>
## PAS (valor numérico)<br>
## PAD (valor numérico)<br>

```python
idade = int(input('Qual a idade? '))
hereditariedade = input('''Qual a história familiar?
Digite 1 para 1 parente com + de 60 anos com doença cardiovascular
Digite 2 para 2 parentes com + de 60 anos com doença cardiovascular
Digite 3 para 1 parente com - de 60 anos com doença cardiovascular
Digite 4 para 2 parentes com - de 60 anos com doença cardiovascular
Digite 5 para 3 parentes ou mais com - de 60 anos com doença cardiovascular ''')
peso = int(input('Qual o peso? '))
altura = float(input('Qual a altura? '))
fumo = int(input('Qual a quantidade de cigarros por dia? '))
exercicio = input('''Qual o nível de atividade física?
Digite 1 para: Esforço Ocupacional e recreacional intenso'
Digite 2 para: Esforço Ocupacional e recreacional moderado'
Digite 3 para: Trabalho sedentário e esforço recreacional intenso'
Digite 4 para: Trabalho sedentário e esforço recreacional moderado'
Digite 5 para: Trabalho sedentário e esforço recreacional leve'
Digite 6 para: Ausência completa de qualquer exercício ''')
gordura_dieta = int(input('Qual a porcentagem de gordura na dieta? '))
pas = int(input('Qual o valor de PAS? '))
pad = int(input('Qual o valor de PAD? '))

registro(idade, hereditariedade, peso, altura, fumo, exercicio, gordura_dieta, pas, pad)
```
